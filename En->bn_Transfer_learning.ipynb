```python
from nltk.translate.bleu_score import corpus_bleu

# Function to calculate accuracy
def calculate_word_accuracy(reference_translations, translated_sentences):
    total_correct_words = 0
    total_words = 0

    for reference, translation in zip(reference_translations, translated_sentences):
        ref_tokens = reference.split()
        trans_tokens = translation.split()

        # Ensure both sentences have the same number of words
        if len(ref_tokens) == len(trans_tokens):
            total_correct_words += sum(1 for ref, trans in zip(ref_tokens, trans_tokens) if ref == trans)
            total_words += len(ref_tokens)


    # Calculate accuracy
    accuracy = total_correct_words / total_words if total_words > 0 else 0
    return accuracy

# Calculate word accuracy
word_accuracy = calculate_word_accuracy(bengali_sentences, ts)

# Calculate BLEU score
bleu_score = corpus_bleu([[reference] for reference in bengali_sentences], ts)

print("Accuracy:", word_accuracy)
print("BLEU Score:", bleu_score)
```

```python
def calculate_accuracy(reference_translations, translated_sentences):
    correct_count = sum(1 for ref, trans in zip(reference_translations, translated_sentences) if ref == trans)
    total_count = len(reference_translations)
    accuracy = correct_count / total_count
    return accuracy
print("Accuracy:",calculate_accuracy(bengali_sentences,ts))
```

```python
def calculate_precision_recall(reference_translations, translated_sentences):
    total_translated = len(translated_sentences)
    total_reference = len(reference_translations)

    # Count correctly translated sentences
    correct_translations = sum(1 for trans in translated_sentences if trans in reference_translations)

    # Calculate precision
    precision = correct_translations / total_translated if total_translated != 0 else 0

    # Calculate recall
    recall = correct_translations / total_reference if total_reference != 0 else 0

    return precision, recall

# Example usage
precision, recall = calculate_precision_recall(bengali_sentences, ts)
print("Precision:", precision)
print("Recall:", recall)
```
